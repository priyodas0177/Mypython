```python
def get_num(marks):
    if marks >=90:
        return 'A+'
    elif marks >=85:
        return "A"
    elif marks >=80:
        return "B+"
    elif marks >=75:
        return "B"
    elif marks >70:
        return "C+"
    else:
        print("fail")

marks=int(input("Enter a number:"))
grade=get_num(marks)
print("The Grade is :",grade)    
            
        
```
